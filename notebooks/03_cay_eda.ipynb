```python
splited_anomaly_rows = [a.split(";") for a in df1.anomaly.to_list() if a is not None]
```

```python
anomalies = []
for row in splited_anomaly_rows:
    anomalies.extend([r.strip() for r in row])
```

```python
anomalies_counter = Counter(anomalies)
```

```python
anomalies_counter
```

```python
categories_split = [row.split(" - ") for row in anomalies]
```

```python
categories_split
```

```python
categories = []
for row in categories_split:
    categories.extend([r.strip() for r in row])
```

```python
categories_counter = Counter(categories)
```
